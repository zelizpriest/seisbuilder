```python
import matplotlib.pyplot as plt
import seisbuilder
```

```python
for i in range(10):
    cube = seisbuilder.SyntheticCube()
    cub, hor, imp = cube.build_seismic()

    fig, ax = plt.subplots(1, 3)
    fig.dpi = 200
    ax[0].imshow(cub[:, 100, :], cmap='seismic')
    ax[0].set_title('Seismic', fontsize=6)
    ax[1].imshow(imp[:, 100, :])
    ax[1].set_title('Impedance', fontsize=6)
    ax[2].imshow(hor[:, 100, :])
    ax[2].set_title('Horizons', fontsize=6)
    for a in ax:
        a.set_xticks([])
        a.set_yticks([])
    plt.subplots_adjust()
    plt.show()
```
